```python
plt.figure(figsize=(14,7))
plt.semilogy(freqs_1420, SDR0_1420_0)
plt.title('Hydrogen Bump, SDR0', size=20)
plt.xlabel('Frequency [Hz]')
plt.ylabel('Power [arb]')
plt.xlim(1420.1, 1421.2)
plt.ylim(10,10**6)
plt.grid()
plt.show()
```
