```python
SDR0 = np.load('FFC_4/data/testing_041624/testing_SDR0_0.npz')['data']
SDR1 = np.load('FFC_4/data/testing_041624/testing_SDR1_0.npz')['data']
```

```python
print((SDR0.shape))
print(len(SDR1))
```

```python
frequency_axis = np.fft.fftshift(np.fft.fftfreq(len(SDR0), 1/3.2))
LO_1420 = 1420
LO_1420p8 = 1420.8
freqs_1420 = frequency_axis + LO_1420
freqs_1420p8 = frequency_axis + LO_1420p8
```

```python
plt.figure(figsize=(14,7))

plt.semilogy(freqs_1420, SDR0, label='SDR0')
plt.semilogy(freqs_1420, SDR1, label='SDR1')
plt.title('HI Line', size=20)
plt.xlabel('Frequency [MHz]', size=15)
plt.ylabel('Power [arb]', size=15)
plt.xlim(1420.1, 1421.2)
plt.ylim(10,10**6)
plt.grid()
plt.legend()
plt.savefig('HI.jpg')
plt.show()
```

```python
plt.figure(figsize=(14,7))
plt.semilogy(freqs_1420, SDR1)
plt.title('HI Line, SDR1', size=20)
plt.xlabel('Frequency [MHz]', size=15)
plt.ylabel('Power [arb]', size=15)
plt.xlim(1420.1, 1421.2)
plt.ylim(40,14**4)
plt.grid()
plt.show()
```
