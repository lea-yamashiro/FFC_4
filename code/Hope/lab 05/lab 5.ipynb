```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def Mandelbrot(z,C):
    zNew = z**2 + C
    return zNew
```

```python
nmax = 30
C_list = [-4,-2.001,-2,-1.999,-1,0,0.5,1,2]

for C in C_list:
    zNew = 0
    print('For C=', C, 'the z values are:')
    for i in range(0,nmax):
        zNew = Mandelbrot(zNew,C)
        print(zNew)
        if abs(zNew)>50:
            print('and the function diverges')
            print()
            break
```

```python
C_values = np.arange(-4,2,0.1)
diverges = np.zeros(len(C_values))

def f(C):
    zNew = 0
    for i in range(nmax):
        zNew = Mandelbrot(zNew,C)
        if abs(zNew)>50:
            return 1
    return 0

for i in range(len(C_values)):
    diverges[i]=f(C_values[i])
    
plt.plot(C_values, diverges)
plt.xlabel('C')
plt.ylabel('f(C)')
plt.show()
```

```python
N = 201
x_values = np.linspace(-2.0, 0.5, N)
y_values = np.linspace(-1.25, 1.25, N)
nmax = 50

grid = np.array([x + y*1j for x in x_values for y in y_values])

A = np.zeros((N, N))

for i in range(N):
    for j in range(N):
        C = grid[i*N + j]
        z = 0
        for n in range(nmax):
            if abs(z) > 2.0:
                A[i, j] = 1
                break
            z = Mandelbrot(z, C)

plt.imshow(A, extent=[-2.0, 0.5, -1.25, 1.25])
plt.colorbar()
plt.title('Mandelbrot Set')
plt.xlabel('Real')
plt.ylabel('Imaginary')
plt.show()
```

```python
N = 250  
nmax = 250  
x_start, x_end, y_start, y_end = -0.75, -0.74, 0.13, 0.14  

x_values = np.linspace(x_start, x_end, N)
y_values = np.linspace(y_start, y_end, N)
grid = np.array([x + y*1j for x in x_values for y in y_values])

A = np.full((N, N), nmax, dtype=int)

for i in range(N):
    for j in range(N):
        C = grid[i*N + j]
        z = 0
        for n in range(nmax):
            if abs(z) > 2.0:
                A[i, j] = n
                break
            z = Mandelbrot(z, C)

min = A.min()
A -= min
max = A.max()
while (True):
    A = (A + 2) % max;
    plt.imshow(A,cmap='gist_ncar', interpolation='nearest',animated=True) 
    plt.draw()
    plt.pause(0.2)
    
```
