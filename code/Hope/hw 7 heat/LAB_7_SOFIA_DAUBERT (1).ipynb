```python
from matplotlib.animation import FFMpegWriter
metadata = dict(title='room', artist='sofia daubert', comment='part 2!')
writer = FFMpegWriter(fps=15, metadata=metadata)
fig = plt.figure()

N = 60
L = 1.0
dx = L / (N - 1)

T = np.zeros((N+2, N+2))
T[1:N+1, 0] = 100
num_iterations = 30

X, Y = np.meshgrid(np.linspace(0, L, N+2), np.linspace(0, L, N+2))

with writer.saving(fig, "part2.mp4", dpi=200):
    for iteration in range(num_iterations):
        Tnew = np.copy(T)

        for i in range(1, N+1):
            for j in range(1, N+1):
                Tnew[i, j] = 0.25 * (T[i, j+1] + T[i, j-1] + T[i-1][j] + T[i+1][j])
        T = Tnew
        
        fig.clear()
        ax = fig.add_subplot(111, projection='3d')
        ax.plot_surface(X, Y, T, cmap=cm.coolwarm, antialiased=False)
        plt.title(f'Iteration {iteration}')
        plt.xlabel('X')
        plt.ylabel('Y')
        plt.pause(0.05)
        writer.grab_frame()
```
