```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def Mandelbrot(z,C):
    zNew = z**2 + C
    return zNew
```

```python
N = 250 
nmax = 250  
x_start, x_end, y_start, y_end = -0.5, 0.5, -0.5, 0.5   

x_values = np.linspace(x_start, x_end, N)
y_values = np.linspace(y_start, y_end, N)
grid = np.array([x + y*1j for x in x_values for y in y_values])

A = np.full((N, N), nmax, dtype=int)

C = complex(-0.4, 0.6) 
for i in range(N):
    for j in range(N):
        z_0 = grid[i * N + j]
        z=z_0
        for n in range(nmax):
            if abs(z) > 2.0:
                A[i, j] = n
                break
            z = Mandelbrot(z, C)

min = A.min()
A -= min
max = A.max()

plt.imshow(A, cmap='gist_ncar', interpolation='nearest')
plt.savefig("mandelbrot_a.png")
plt.show()
```
