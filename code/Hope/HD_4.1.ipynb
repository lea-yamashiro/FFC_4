```python
frequency_axis = np.fft.fftshift(np.fft.fftfreq(len(SDR0), 1/3.2))
LO_1420 = 1420
LO_1420p8 = 1420.8
freqs_1420 = frequency_axis + LO_1420
freqs_1420p8 = frequency_axis + LO_1420p8
```

```python
plt.figure(figsize=(14,7))
plt.semilogy(freqs_1420, SDR0)
plt.title('HI Line, SDR0', size=20)
plt.xlabel('Frequency [MHz]')
plt.ylabel('Power [arb]')
plt.xlim(1420.1, 1421.2)
plt.ylim(10,10**6)
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(14,7))
plt.semilogy(freqs_1420, SDR1)
plt.title('HI Line, SDR1')
plt.xlabel('Frequency [MHz]')
plt.ylabel('Power [arb]')
plt.xlim(1420.1, 1421.2)
plt.ylim(40,14**4)
plt.grid()
plt.show()
```
