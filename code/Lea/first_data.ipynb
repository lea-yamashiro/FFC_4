```python
frequency_axis = np.fft.fftshift(np.fft.fftfreq(len(SDR0_1420_0), 1/3.2))
LO_1420 = 1420
LO_1420p8 = 1420.8
freqs_1420 = frequency_axis + LO_1420
freqs_1420p8 = frequency_axis + LO_1420p8
```

```python
plt.semilogy(freqs_1420, SDR0_1420_0, label = 'LO 1420')
plt.title('Hydrogen Bumpin, SDR0')
plt.xlabel('Frequency [Hz]')
plt.ylabel('Power [arb]')
plt.legend(facecolor='white')
#plt.xlim(1420.1, 1421.2)
#plt.ylim(10,10**6)
```

```python
plt.semilogy(freqs_1420, SDR0_1420_0, label = 'SDR0')
plt.semilogy(freqs_1420, SDR1_1420_0, label = 'SDR1', color = 'darkorange')
plt.title('Hydrogen Bumpin, Both SDRs')
plt.xlabel('Frequency [Hz]')
plt.ylabel('Power [arb]')
plt.xlim(1420.1, 1421.2)
plt.ylim(20,19**4)
plt.legend(facecolor='white')
```
